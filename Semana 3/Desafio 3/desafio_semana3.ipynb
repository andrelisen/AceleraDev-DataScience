### Objetivo

Queremos conhecer melhor nossos clientes por estado. Para isso, iniciamos uma análise na pontuação de crédito. 
Para realizar a verificação inicial, precisamos de alguns valores.
Os valores são a média, a mediana, a moda e o desvio padrão da pontuação de crédito.

```python
import pandas as pd
import json
```

```python
df = pd.read_csv('desafio1.csv')
```

```python
df
```

```python
df.dtypes.value_counts()
```

```python
df.isna().sum()
```

```python
df['estado_residencia'].unique()
```

O que preciso calcular?
* Média
* Mediana
* Moda
* Desvio Padrão

### Média

```python
df.groupby('estado_residencia')['pontuacao_credito'].mean()
```

```python
medias = df.groupby('estado_residencia')['pontuacao_credito'].mean()
```

### Mediana


```python
df.groupby('estado_residencia')['pontuacao_credito'].median()
```

```python
medianas = list(df.groupby('estado_residencia')['pontuacao_credito'].median())
```

```python
print(medianas[0])
```

### Moda

```python
moda_PR = list(df.groupby(df['estado_residencia'] == 'PR')['pontuacao_credito'].value_counts().to_dict())
```

```python
print(moda_PR[0][1])
```

```python
moda_RS = list(df.groupby(df['estado_residencia'] == 'RS')['pontuacao_credito'].value_counts().to_dict())
```

```python
print(moda_RS[0][1])
```

```python
moda_SC = list(df.groupby(df['estado_residencia'] == 'SC')['pontuacao_credito'].value_counts().to_dict())
```

```python
print(moda_SC[0][1])
```

### Desvio padrão 

```python
df.groupby('estado_residencia')['pontuacao_credito'].std()
```

```python
desvios = df.groupby('estado_residencia')['pontuacao_credito'].std()
```

### Escrever em um arquivo JSON

```python
resposta_desafio = {"SC": {"moda": moda_SC[0][1], "mediana": medianas[2], "media": medias[2], "desvio_padrao": desvios[2]},
                    "RS": {"moda": moda_RS[0][1], "mediana": medianas[1], "media": medias[1], "desvio_padrao": desvios[1]},
                    "PR": {"moda": moda_PR[0][1], "mediana": medianas[0], "media": medias[0], "desvio_padrao": desvios[0]}
                   }


```

```python
with open("submission.json", "w") as json_file:
    json.dump(resposta_desafio, json_file)
```
